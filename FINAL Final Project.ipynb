The symptom list has been updated to cough, shortness of breath, chills, fever, headache, sore throat, loss of taste or smell, shaking associated with chills, and/or body aches. ## Import COVID-19 Data

```python
import pandas
```

```python
df = pandas.read_excel('s3://ward-ia241/COVID-19-geographic-disbtribution-worldwide-2020-04-28.xlsx')
df
```

## Data Visualizations and Calculations

### What Continent Has the Highest Death Count?

```python
death_rate_by_continent = df.groupby('continentExp').mean()['deaths']
death_rate_by_continent.plot.bar()
```

The America's have the highest amount of deaths, with Europe close behind.

### What are the Top Ten Countries with the Highest Infection Count?

```python
sum_infection_by_country = df.groupby('countriesAndTerritories').sum()['cases']
sum_infection_by_country.nlargest(10).plot.bar()
```

Spain has the second highest amount of cases, beat out only by the United States of America.

### Relation Between Number of Cases and Number of Deaths in Spain

```python
spain_data = df.loc[df['countriesAndTerritories'] == 'Spain']
spain_data[:10]
```

```python
spain_data.plot.scatter(x='cases',y='deaths')
```

Positive relationship with the amount of cases and the amount of deaths. As the number of cases raise the number of deaths do as well.

### Timeline of Spain's Infection Cases by Month (January 2020 - April 2020)

```python
spain_sum_death_by_date = spain_data.groupby('dateRep').sum()['cases']

spain_sum_death_by_date.plot()
```

There have been spikes in case amounts for Spain in the month of March, decreasing overall in April.
