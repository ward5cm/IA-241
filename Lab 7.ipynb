```python
from collections import Counter
```

```python
with open('IA_jobs.txt','r') as IA_jobs:
    word_list = IA_jobs.read().split()
    
    count_result = Counter(word_list)
    for word,count in count_result.most_common(20):
        print(word,count)
```

### Q3

```python
import xlwt
```

```python
book = xlwt.Workbook()

sheet = book.add_sheet ('word_count')
```

```python
i = 0

sheet.write(i,0,'word')
sheet.write(i,1,'count')
```

```python
with open ('IA_jobs.txt','r') as IA_jobs:
    count_result = Counter(IA_jobs.read().split())
    
    for word,count in count_result.most_common(20):
        
        i = i +1
        sheet.write(i,0,word)
        sheet.write(i,1,count)
        
```

```python
book.save('Lab_7_word_count.xls')
```
